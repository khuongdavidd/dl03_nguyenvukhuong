```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from my_lib import *
import pandas as pd
from datetime import datetime
import re
```

```python
option = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=option)
driver.maximize_window()

```

# lấy link từng sản phẩm

```python


# Mở trang web
driver.get('https://www.coursera.org/search?query=data%20science&language=English&topic=Data%20Science&sortBy=BEST_MATCH')

# Tìm phần tử mà bạn muốn cuộn tới
element = driver.find_element(By.CLASS_NAME, 'page-footer-container')
while True:
# Cuộn trang đến vị trí trên đầu phần tử một khoảng cách nhỏ (ví dụ: 100 pixels)
    offset = -700  # Khoảng cách cuộn lên trên phần tử
    driver.execute_script("window.scrollTo(0, arguments[0] - arguments[1]);", element.location['y'], -offset)
    time.sleep(1)  # Đợi 1 giây sau mỗi lần cuộn trang
    tag_cards = driver.find_elements(By.CSS_SELECTOR, 'div.cds-ProductCard-gridCard')
    if len(tag_cards) > 1387:
        break
        print("số tag card: ", len(tag_cards))
    else:
        print("số tag card: ", len(tag_cards))
```

kiểm tra lại

```python
print("số tag card: ", len(tag_cards))
```

tìm link khóa học

```python
list_link = []

for tag_card in tag_cards:
    tag_link = tag_card.find_element(By.TAG_NAME, 'a')
    link = tag_link.get_attribute('href')
    list_link.append(link)
```

ghi vào file

```python
print(len(list_link))
write_txt('data/txt/link_course.txt', list_link)
```

Lấy list link

```python
list_links = read_txt('data/txt/link_course.txt')
print(len(list_links))
```

khởi chạy vòng for

```python
i = 0
list_course_id = []
list_course_name = []
list_unit = []
list_review_number = []
list_avg_star = []
list_level = []
list_result = []

# Khởi tạo trình điều khiển trình duyệt (ở đây là Chrome)
driver = webdriver.Chrome()

for link in list_links:
    try:
        driver.get(link)
        time.sleep(1)
        i += 1

        # lấy đơn vị phát hành
        tag_unit = driver.find_element(By.XPATH, '//div[@class="css-1qiyodj"]/img')
        unit = tag_unit.get_attribute('alt')

        # lấy id khóa học
        tag_course_id = driver.find_element(By.TAG_NAME, 'h1')
        course_id = tag_course_id.text

        # lấy tag review
        tag_review = driver.find_element(By.XPATH, '//section[@class="css-3nq2m6"]/div[@class="css-lt1dx1"]')

        # lấy đánh giá
        tag_review_number = tag_review.find_element(By.XPATH, value='//*[@id="rendered-content"]/div/main/section[2]/div/div/div[2]/div/div/section/div[2]/div[1]/p')
        review_number_text = tag_review_number.text
        review_number = int(re.sub(r'\D', '', review_number_text))


        # lấy sao trung bình
        tag_avg_star = tag_review.find_element(By.XPATH, '//*[@id="rendered-content"]/div/main/section[2]/div/div/div[2]/div/div/section/div[2]/div[1]/div[1]')
        avg_star = eval(tag_avg_star.text)


        # lấy level
        tag_level = tag_review.find_element(By.XPATH, value='//*[@id="rendered-content"]/div/main/section[2]/div/div/div[1]/div[2]/section/div[2]/div[2]/div[1]')
        level = tag_level.text


        # lấy kết quả
        tag_result = driver.find_element(By.CSS_SELECTOR, 'ul.cds-9')
        result = tag_result.text


    except Exception as e:
        print(str(e))
        print(f"{i} thất bại tại link {link}")
    else:
        list_course_id.append(course_id)
        list_course_name.append(course_id)
        list_review_number.append(review_number)
        list_avg_star.append(avg_star)
        list_unit.append(unit)
        list_result.append(result)
        list_level.append(level)
        print(f"{i} hoàn thành link {link}")

# Đóng trình điều khiển
driver.quit()
```

```python
print(len(list_review_number), len(list_unit), len(list_result), len(list_avg_star), len(list_course_id), len(list_course_name), len(list_level))
data = {
    'Course ID': list_course_id,
    'Course Name': list_course_name,
    'Unit': list_unit,
    'Level': list_level,
    'Avg Star': list_avg_star,
    'Review Number': list_review_number,
    'Result': list_result
}
```

```python
df = pd.DataFrame(data=data)
df
```

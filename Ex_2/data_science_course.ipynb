```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from my_lib import *
import pandas as pd
from datetime import datetime
```

```python
option = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=option)
driver.maximize_window()

```

# lấy link từng sản phẩm

```python
list_links = []
for i in range(1, 85):
    url = f"https://www.coursera.org/search?query=data%20science&language=English&topic=Data%20Science&page={i}&sortBy=BEST_MATCH"
    driver.get(url)
    # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)
    course_container = driver.find_element(by=By.CSS_SELECTOR, value='ul.css-reop8o')
    course_links =  course_container.find_elements(by=By.TAG_NAME, value='a')

    for course_link in course_links:
        link = course_link.get_attribute('href')
        list_links.append(link)
    print(f"hoàn thành trang {i}")
else:
    write_txt('data/txt/link_course.txt', list_links)


```

Lấy list link

```python
list_links = read_txt('data/txt/link_course.txt')
print(len(list_links))
```

khởi chạy vòng for

```python
i = 0
list_course_id = []
list_course_name = []
list_unit = []
list_review_number = []
list_avg_star = []
list_level = []
list_result = []
for link in list_links:
    try:
        driver.get(link)
        time.sleep(3)
        i+=1
        
        # lấy đơn vị phát hành
        tag_unit = driver.find_element(by=By.XPATH, value='//div[@class="css-1qiyodj"]/img')
        unit = tag_unit.get_attribute('alt')


        #lấy id khóa học
        tag_course_id = driver.find_element(by=By.TAG_NAME, value='h1')
        course_id = tag_course_id.text

        # lấy tag review
        tag_review = driver.find_element(by=By.XPATH, value='//section[@class="css-3nq2m6"]/div[@class="css-lt1dx1"]')
        
        #lấy đánh giá
        tag_review_number = tag_review.find_element(by=By.CSS_SELECTOR, value='p.css-vac8rf')
        review_number = int(''.join(filter(str.isdigit, tag_review_number)))

        #lấy sao trung bình
        
        avg_star = eval(tag_review.text.split('\n')[0])

        #lấy level
        tag_level = driver.find_element(by=By.XPATH, value='//section[@class="css-3nq2m6"]/div[@class="css-lt1dx1"]/div[2]/div[1]')
        level = tag_review.text.split('\n')[2]


        #lấy kết quả
        tag_result = driver.find_element(by=By.CSS_SELECTOR, value='ul.cds-9')
        result = tag_result.text
    except Exception as e:
        print(str(e))
        print(f"{i} thất bại tại link {link}")
    else:
        list_course_id.append(course_id)
        list_course_name.append(course_id)
        list_review_number.append(review_number)
        list_avg_star.append(avg_star)
        list_unit.append(unit)
        list_result.append(result)
        list_level.append(level)
        print(f"{i} hoàn thành link {link}")
```

```python
print(len(list_review_number), len(list_unit), len(list_result), len(list_avg_star), len(list_course_id), len(list_course_name), len(list_level))
data = {
    'Course ID': list_course_id,
    'Course Name': list_course_name,
    'Unit': list_unit,
    'Level': list_level,
    'Avg Star': list_avg_star,
    'Review Number': list_review_number,
    'Result': list_result
}
```

```python
df = pd.DataFrame(data=data)
df
```
